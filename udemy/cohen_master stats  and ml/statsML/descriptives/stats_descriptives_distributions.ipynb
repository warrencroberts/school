```python
def number_of_character_occurrences(word, char):
    return len([x for x in word.lower() if x == char.lower()])

def find_word_with_max_chars(words, char):
    max_char_count = 0
    max_word = ""
    for word in words:
        curr_len = number_of_character_occurrences(word, char)
        if curr_len > max_char_count:
            max_char_count = curr_len
            max_word = word
    return max_word

char_to_count = 'i'
print(f"Word with the most {char_to_count}'s: {find_word_with_max_chars(['Mississippi', 'Louisiana'], char_to_count)}")
```

```python
def possible_palindrome(pal_prospect):
    """
    Function returns True if pal_prospect may be turned into a palindrome and valse otherwise
    :param string pal_prospect: The string to test
    :return True if string may be palidrome or False otherwise
    
    A Palindrome 
        Must contain an odd number of characters.
        Must contain one character repeated an odd number of times
        All other characters must be repeated an even number of times
    """
    str_len = len(pal_prospect) 
    if ((str_len % 2) == 0) or (str_len < 3):
        print(f'Even number of characters: {str_len}')
        return False

    saw_odd = False
    for char in ''.join(set(pal_prospect)):
        occurrences = number_of_character_occurrences(pal_prospect, char)
        print(f'char: {char} occurrences: {occurrences}')
        if (occurrences % 2) != 0:
            if saw_odd:
                return False
            else:
                saw_odd = True
    return saw_odd

possible_palindrome('acbadbaca')
```

```python
## Normally-distributed random numbers

# parameters
stretch = 1 # variance (square of standard deviation)
shift   = 5 # mean
n       = 1000

# create data
data = stretch*np.random.randn(n) + shift

# plot data
plt.hist(data,25)
plt.title('Empirical normal distribution')
plt.show()
```

```python
## Uniformly-distributed numbers

# parameters
stretch = 2 # not the variance
shift   = .5
n       = 10000

# create data
data = stretch*np.random.rand(n) + shift-stretch/2

# plot data
fig,ax = plt.subplots(2,1,figsize=(5,6))

ax[0].plot(data,'.',markersize=1)
ax[0].set_title('Uniform data values')

ax[1].hist(data,25)
ax[1].set_title('Uniform data histogram')

plt.show()
```

```python
## log-normal distribution

N = 1001
x = np.linspace(0,10,N)
lognormdist = stats.lognorm.pdf(x,1)

plt.plot(x,lognormdist)
plt.title('Analytic log-normal distribution')
plt.show()
```

```python
## empirical log-normal distribution

shift   = 5  # equal to the mean?
stretch = .5 # equal to standard deviation?
n = 2000     # number of data points

# generate data
data = stretch*np.random.randn(n) + shift
data = np.exp( data )

# plot data
fig,ax = plt.subplots(2,1,figsize=(4,6))
ax[0].plot(data,'.')
ax[0].set_title('Log-normal data values')

ax[1].hist(data,25)
ax[1].set_title('Log-normal data histogram')
plt.show()
```

```python
## binomial

# a binomial distribution is the probability of K heads in N coin tosses,
# given a probability of p heads (e.g., .5 is a fair coin).

n = 10 # number on coin tosses
p = .5 # probability of heads

x = range(n+2)
bindist = stats.binom.pmf(x,n,p)

plt.bar(x,bindist)
plt.title('Binomial distribution (n=%s, p=%g)'%(n,p))
plt.show()
```

```python
## t

x  = np.linspace(-4,4,1001)
df = 200
t  = stats.t.pdf(x,df)

plt.plot(x,t)
plt.xlabel('t-value')
plt.ylabel('P(t | H$_0$)')
plt.title('t(%g) distribution'%df)
plt.show()
```

```python
## F

# parameters
num_df = 5   # numerator degrees of freedom
den_df = 100 # denominator df

# values to evaluate 
x = np.linspace(0,10,10001)

# the distribution
fdist = stats.f.pdf(x,num_df,den_df)

plt.plot(x,fdist)
plt.title(f'F({num_df},{den_df}) distribution')
plt.xlabel('F value')
plt.show()
```
