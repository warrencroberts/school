```python
def gen_outliers(mean, std, magnitude, number):
    rand_num = np.random.rand(number)
#     rand_sign= np.random.choice([-1, 1], number)
    rand_sign = +1
    
    return (rand_num + (mean + (std * magnitude))) * rand_sign
```

```python
def histo_with_outliers(data, magnitude, number):
    outliers = gen_outliers(data.mean(), data.std(), magnitude, number)
#     print(outliers)
    data_with_outliers = np.append(data, outliers)
    print(f"Original Mean:{data.mean()}, Original Median:{statistics.median(data)}")
    print(f"New Mean     :{data_with_outliers.mean()},  New Median.    :{statistics.median(data_with_outliers)}\n\n")

#     plt.figure(figsize=(16, 6))
#     ax1 = sns.histplot(data, kde=True)
#     ax1.set_title('Original Distribution')
#     ax1.set_xlabel('Z score')
#     ax1.set_ylabel('Count')
#     plt.show()
    
    ax2 = sns.histplot(data_with_outliers, kde=True)
    ax2.figure.set_figheight(6)
    ax2.figure.set_figwidth(20)
    ax2.set_title('New Distribution')
    ax2.set_xlabel('Z score')
    ax2.set_ylabel('Count')
    plt.show()
```

```python
histo_with_outliers(small_data, 5, 10)
```

```python
histo_with_outliers(small_data, 150, 10)
```

```python
histo_with_outliers(small_data, 1500, 10)
```
