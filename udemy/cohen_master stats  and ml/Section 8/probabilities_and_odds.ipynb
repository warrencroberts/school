```python
import numpy as np
import seaborn as sns
```

```python
def probability_from_odds(n, m):
    return 1.0 / (1.0 + ((n+1)/(m+1)))
```

```python
def plot_heatmap(n, m):
    range_n = range(n)
    range_m = range(m)

    probabilities = np.zeros((n, m))
    for n in range_n:
        for m in range_m:
            probabilities[n][m] = probability_from_odds(n,m)
            
    ax = sns.heatmap(probabilities,annot=True)
    ax.figure.set_figwidth(25)
    ax.figure.set_figheight(18)
    ax.set_xlabel('m -- For')
    ax.set_ylabel('n -- Against')
    ax.set_xticklabels(range(1,m+2))
    ax.set_yticklabels(range(1,n+2))
```

```python
plot_heatmap(10,11)
```

```python
plot_heatmap(10,15)
```
