```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import default_rng
import scipy.stats as stats
import seaborn as sns
```

```python
## generate the data

# parameters
N = 20  # sample size
popMu = .5 # true population mean
rng = np.random.default_rng()
sigma = 1.0
data = rng.normal(popMu, sigma, N)

# let's see what the data look(s) like
plt.plot(data,'ko',markerfacecolor='w',markersize=10)
plt.xlabel('Data index')
plt.ylabel('Data value')
plt.show()
```

```python
the_range = np.arange(0.1, 3.1, 0.001)
t_stats = np.zeros(len(the_range))
p_vals  = np.zeros(len(the_range))
for i,sigma in enumerate(the_range):
    data = rng.normal(popMu, sigma, N)
    t_stats[i], p_vals[i] = stats.ttest_1samp(data,0)

fig, axs = plt.subplots(2, figsize=(20, 12))
fig.suptitle('Standard Deviation affects on t-statistic p-value')

sns.color_palette("OrRd")
sns.set(rc={'axes.facecolor':'bisque', 'figure.facecolor':'oldlace'})

sns.scatterplot(ax=axs[0], x=the_range, y=t_stats)
axs[0].set_ylabel('t-statistic')
axs[0].legend('t-statistic')

plt.xlabel('Standard Deviation')

sns.scatterplot(ax=axs[1], x=the_range, y=p_vals)
axs[1].set_ylabel('p-value')
```

```python
fig = plt.figure(figsize=(20, 10))
fig.suptitle('t-statistic and p-value')

sns.color_palette("OrRd")
sns.set(rc={'axes.facecolor':'bisque', 'figure.facecolor':'oldlace'})

axs = sns.scatterplot(x=t_stats, y=p_vals)
axs.set_ylabel('p-value')
plt.xlabel('')
```
