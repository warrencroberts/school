## Categorical Plot

### Count Plot

```python
titanic['sex'].unique()
titanic['class'].unique()
```

```python
sns.countplot(x='sex',  data= titanic)
```

```python
sns.stripplot(x='sex', y='total_bill', hue='day', data=tips, jitter=True)
```

```python
sns.swarmplot(x='sex', y='total_bill', data=tips)
```

```python
sns.violinplot(x='sex', y='total_bill', data=tips)
```

```python
#sns.barplot(x='sex', y ='age', data=titanic)
sns.barplot(x='sex', y ='age', hue='class',  data=titanic, order=['female', 'male'], estimator=np.std, palette='Set1')
```

```python
sns.boxplot(x='sex', y='age',data=titanic)
```
