### Functional method - Multiple Subplot

```python
#plt.plot(x,y,x,z)
plt.plot(x,y)
plt.plot(x,z)
```

```python
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'b')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('1st Chart')

plt.subplot(1,2,2)
plt.plot(y,x, 'r')
plt.xlabel('Y')
plt.ylabel('X')
plt.title('2nd Chart')

plt.tight_layout()
```

### Object Oriented way - Single Plot

```python
#fig = plt.figure()
#ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
#ax.plot(x,y)

fig = plt.figure(figsize=(10,5))
ax = fig.add_axes([0.1, 0.1, 0.5, 0.9])
ax.plot(x,y, 'b')
```

### Object Oriented way - Overlapping  Plot

```python
fig= plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax2 = fig.add_axes([0.4, 0.65, 0.25, 0.25])

ax1.plot(x,y, color= 'magenta')
ax2.plot(x,y, color = 'green')

ax1.set_title('Bigger graph')
ax2.set_title('Smaller graph')
```

### Object Oriented way - Sub  Plot

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize = (10,5))

ax[0].plot(x,y)
ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')
ax[0].set_title('1st Chart')

ax[1].plot(x,y)
ax[1].set_xlabel('X')
ax[1].set_ylabel('Y')
ax[1].set_title('2nd Chart')

plt.tight_layout()
```

## Color & Marker

```python
plt.style.use('ggplot')
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax.plot(x,y, c= 'r')
ax.plot(x,y+1, c= 'g')
ax.plot(x,y+2, c= 'b')
ax.plot(x,y+3, c= '#FFFF00')
ax.plot(x,y+4, c= '#FF00FF')
ax.plot(x,y+5, c= 'r', marker = 'o', markersize = 5, lw = 2.5)
ax.plot(x,y+6, c= 'b', marker = 'D', markersize = 5, lw = 0.5)
```

## Add legend

```python
fig = plt.figure()
ax = fig.add_axes([0,0,0.9,0.9])
ax.plot(x, y, label='Sin')
ax.plot(x, z, label= 'Cosine')
ax.legend(loc=0)
```

## Ticks

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax.set_xticks([1,3,5,7,9])
ax.set_yticks([-0.6, 0.1, 0.6])
ax.plot(x,y)
```

## Step Plot & Fill Area

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax.step(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax.fill_between(x, x**2, x**3, color ='orange')
```

## Grid

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax.grid(color ='b', linestyle='solid')
# 'solid' | 'dashed', 'dashdot', 'dotted'
ax.plot(x,x*x)
```

## Text annotation in Graph

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax.text(2, 60, "$y=x^2$", fontsize= 15, color = 'blue')
ax.plot(x, x**2)
```
