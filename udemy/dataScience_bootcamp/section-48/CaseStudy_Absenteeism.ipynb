```python
import pandas as pd

raw_csv_data = pd.read_csv("Absenteeism-data.csv")
df = raw_csv_data.copy()
df = df.drop(['ID'], axis=1)
df.head()
```

```python
RFA = 'Reason for Absence'
df[RFA].min()
```

```python
df[RFA].max()
```

```python
# number of unique values
len(df[RFA].unique())
```

```python
reason_columns = pd.get_dummies(df[RFA])
reason_columns.head()
```

```python
reason_columns['check'] = reason_columns.sum(axis=1)
reason_columns.head()
```

```python
reason_columns['check'].sum(axis=0)
```

```python
reason_columns['check'].unique()
```

```python
reason_columns = reason_columns.drop(['check'], axis=1)
```

```python
reason_columns = pd.get_dummies(df[RFA], drop_first=True)
reason_columns.head()
```

Split dummies into classes

```python
reason_type_1 = reason_columns.loc[:,1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:,15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:,18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:,22:].max(axis=1)
```

```python
reason_type_4.head()
```

```python
df = df.drop(['Reason for Absence'], axis=1)
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis=1)
```

```python
df.columns.values
```

```python
# df_with_reasons_and_age['Date'] = pd.to_datetime(df_with_reasons_and_age['Date'], '%d/%m/%Y')
df_2['Date'] = pd.to_datetime(df_2['Date'], format = '%d/%m/%Y')

list_months = []
for i in range(df_2.shape[0]):
    list_months.append(df_2['Date'][i].month)
    
df_2['Month Value'] = list_months
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_2['Day of the week'] = df_2.Date.apply(date_to_weekday)
```
