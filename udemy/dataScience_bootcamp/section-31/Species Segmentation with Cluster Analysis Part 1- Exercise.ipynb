# Clustering (unscaled data)

Separate the original data into 2 clusters.

```python
from sklearn import preprocessing

def fit_and_wcss(num_clusters, input_data):
    print("num_clusters: {0}".format(num_clusters))
    x_raw = input_data.copy()
    
    kmeans_raw = KMeans(num_clusters)
    kmeans_raw.fit(x_raw)
    x_raw['Cluster'] = kmeans_raw.fit_predict(x_raw)

    x_std = pd.DataFrame(preprocessing.scale(input_data.copy()),
                        columns=input_data.columns.values)
    kmeans_std = KMeans(num_clusters)
    kmeans_std.fit(x_std)
    x_std['Cluster'] = kmeans_std.fit_predict(x_std)

    fig, axs = plt.subplots(1,2, figsize=(12,6))   
    axs[0].scatter(x_raw['sepal_length'],x_raw['sepal_width'],c=x_raw['Cluster'],cmap='rainbow')
    axs[1].scatter(x_std['sepal_length'],x_std['sepal_width'],c=x_std['Cluster'],cmap='rainbow')

    plt.show()

    return kmeans.inertia_
```

```python
wcss = [fit_and_wcss(x, raw_data) for x in range(1,6)]
```
