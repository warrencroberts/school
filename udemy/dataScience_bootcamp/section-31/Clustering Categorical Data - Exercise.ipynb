## Import the relevant libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'Categorical.csv'</i>.

```python
# Load the data
raw_data = pd.read_csv('Categorical.csv')
# Check the data
raw_data.head()
```

Remove the duplicate index column from the dataset. 

```python
data = raw_data.copy()
data = data.drop(['name'], axis = 1)
data.iloc[:,0:3].head()
```

```python
data['continent'].unique()
```

## Map the data

Use the <i>'continent'</i> category for this analysis.

```python
data['continent'] = data['continent'].map({'North America':0, 
                                                 'Asia': 1,
                                                 'Africa':2,
                                                 'Europe':3,
                                                 'South America':4,
                                                 'Oceania':5,
                                                 'Antarctica':6,
                                                 'Seven seas (open ocean)': 7})
```

```python
def fit_and_plot(num_clusters, input_data):
    data = input_data.copy()
    x = data.iloc[:,0:3]
    
    kmeans = KMeans(num_clusters)
    kmeans.fit(x)
    
    identified_clusters = kmeans.fit_predict(x)

    data['Cluster'] = identified_clusters

    plt.scatter(data['Longitude'],data['Latitude'],c=data['Cluster'],cmap='rainbow')
    plt.xlim(-180,180)
    plt.ylim(-90,90)
    plt.show()
```

## Clustering

Use 4 clusters initially.

```python
fit_and_plot(2, data)
```

```python
fit_and_plot(3, data)
```

```python
fit_and_plot(4, data)
```

```python
fit_and_plot(7, data)
```

```python
fit_and_plot(8, data)
```
